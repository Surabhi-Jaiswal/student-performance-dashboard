```python
from dash import Dash, dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import dash_bootstrap_components as dbc
```

```python
app = Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])
```

```python
df = pd.read_csv('student_data.csv')
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
numeric_data = df.select_dtypes(include=[float, int])
```

```python
app.layout = dbc.Container([
    dbc.Row([
        dbc.Col(html.H1("Education Performance Dashboard", style={'text-align': 'center'}), width=12)
    ], className="mb-4"),
    
    dbc.Row([
        dbc.Col(html.Div([
            html.H4("Feature Selection"),
            html.P("Select a feature to visualize its distribution, scatter plot, and box plot."),
            dcc.Dropdown(
                id='feature_dropdown',
                options=[{'label': col, 'value': col} for col in df.select_dtypes(include=[float, int]).columns],
                value='G3',  # Default value
                style={'width': '100%'}
            )
        ]), width=4),
        
        dbc.Col(html.Div([
            html.H4("Filter Data"),
            html.P("Select a column to filter and choose values from the dropdown."),
            dcc.Dropdown(
                id='filter_dropdown',
                options=[{'label': col, 'value': col} for col in df.columns],
                value=None,
                multi=False,
                style={'width': '100%'}
            ),
            dcc.Dropdown(
                id='filter_value',
                options=[],
                value=[],
                multi=True,
                style={'width': '100%'}
            ),
        ]), width=4),
    ], className="mb-4"),
    
    dbc.Row([
        dbc.Col(dcc.Graph(id='correlation_heatmap'), width=12)
    ], className="mb-4"),
    
    dbc.Row([
        dbc.Col(dcc.Graph(id='feature_distribution'), width=6),
        dbc.Col(dcc.Graph(id='scatter_plot'), width=6),
    ], className="mb-4"),
    
    dbc.Row([
        dbc.Col(dcc.Graph(id='box_plot'), width=6),
        dbc.Col(dcc.Graph(id='attendance_pie_chart'), width=6),
    ], className="mb-4"),
    
    dbc.Row([
        dbc.Col(dcc.Graph(id='grades_bar_chart'), width=12)
    ], className="mb-4"),
    
    dbc.Row([
        dbc.Col(html.Div(id='motivational_quote', style={'text-align': 'center', 'margin-top': '20px'}), width=12)
    ]),
    
    dbc.Row([
        dbc.Col(html.Div([
            html.H4("Instructions:"),
            html.P("1. Use the 'Feature Selection' dropdown to choose a feature for detailed analysis."),
            html.P("2. Select a column and value from the 'Filter Data' section to apply filters to the data."),
            html.P("3. Visualizations will update automatically based on the selected feature and filter."),
            html.P("4. Check out the motivational quote at the bottom for a bit of inspiration.")
        ]), width=12)
    ], className="mt-4")
])
```

```python
@app.callback(
    Output('filter_value', 'options'),
    [Input('filter_dropdown', 'value')]
)
def set_filter_options(selected_column):
    if selected_column and selected_column in df.columns:
        unique_values = df[selected_column].dropna().unique()
        return [{'label': str(val), 'value': str(val)} for val in unique_values]
    return []

# Update the graphs based on the selected feature and filter
@app.callback(
    [Output('correlation_heatmap', 'figure'),
     Output('feature_distribution', 'figure'),
     Output('scatter_plot', 'figure'),
     Output('box_plot', 'figure'),
     Output('attendance_pie_chart', 'figure'),
     Output('grades_bar_chart', 'figure'),
     Output('motivational_quote', 'children')],
    [Input('feature_dropdown', 'value'),
     Input('filter_dropdown', 'value'),
     Input('filter_value', 'value')]
)
def update_graphs(selected_feature, filter_feature, filter_value):
    # Initialize filtered dataframe
    filtered_df = df.copy()

    # Apply filter if selected
    if filter_feature and filter_value:
        filtered_df = filtered_df[filtered_df[filter_feature].astype(str).isin(filter_value)]

    # Correlation Heatmap
    numeric_data = filtered_df.select_dtypes(include=[float, int])
    correlation_matrix = numeric_data.corr()
    heatmap_fig = px.imshow(correlation_matrix, text_auto=True, aspect="auto")

    # Feature Distribution
    feature_dist_fig = px.histogram(filtered_df, x=selected_feature)

    # Scatter Plot of Grades vs Selected Feature
    scatter_fig = px.scatter(filtered_df, x=selected_feature, y='G3', trendline="ols")

    # Box Plot of Grades by Selected Feature
    box_fig = px.box(filtered_df, x=selected_feature, y='G3')

    # Attendance Pie Chart
    attendance_fig = px.pie(filtered_df, names='school', values='absences')

    # Grades Bar Chart
    grades_fig = go.Figure(data=[
        go.Bar(name='G1', x=filtered_df['school'], y=filtered_df['G1']),
        go.Bar(name='G2', x=filtered_df['school'], y=filtered_df['G2']),
        go.Bar(name='G3', x=filtered_df['school'], y=filtered_df['G3'])
    ])
    grades_fig.update_layout(barmode='group')

    # Random motivational quote
    quotes = [
        "Education is the most powerful weapon which you can use to change the world.",
        "The beautiful thing about learning is that no one can take it away from you.",
        "Education is not preparation for life; education is life itself.",
        "The only limit to our realization of tomorrow is our doubts of today."
    ]
    quote = quotes[0]  # Example: You can modify to choose a random quote based on any criteria.

    return heatmap_fig, feature_dist_fig, scatter_fig, box_fig, attendance_fig, grades_fig, quote
```

```python
app.run_server(debug=True, use_reloader=False, port=8053)
```
